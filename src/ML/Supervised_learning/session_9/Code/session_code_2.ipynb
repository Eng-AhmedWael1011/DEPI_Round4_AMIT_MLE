```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier , RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
```

```python
data = pd.read_csv('../../../data/creditcard.csv')
data.head()
```

```python
x = data.drop(['Class'] , axis=1)
y = data['Class']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,random_state=42 , test_size=0.2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbm_param = {
    'n_estimators': [10,50,100,200],
    'learning_rate': [0.01 , 0.1],
    'max_depth': [3,4,5,6],
    'subsample': [0.7 , 0.8 , 0.9],
    'min_samples_split': [2,5,10]
}

```

```python
gb_grid = GridSearchCV(GradientBoostingClassifier() , param_grid=gbm_param , cv=3, n_jobs=-1 , scoring= 'roc_auc' , verbose=2)

gb_cv = gb_grid.fit(x_train , y_train)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print('The best parameters combination is')
print(gb_cv.best_params_)

# prediction using Model
y_pred_gbm = gb_cv.best_estimator_.predict(y_test)
cm_gbm = confusion_matrix(y_test , y_pred_gbm)
print(cm_gbm)
print(classification_report(y_test , y_pred_gbm , target_names=["Safe" , "Fraud"]))

# calculate sensitivity , specificity, accuracy

total_gbm = sum(cm_gbm)
accuracy_gbm = (cm_gbm[0,0] + cm_gbm[1,1]) / total_gbm
print('Accuracy (GBM): ' , accuracy_gbm)

sensitivity_gbm = cm_gbm[0,0] / (cm_gbm[0,0] + cm_gbm[1,1])
print('Sensitivity (GBM)' , sensitivity_gbm)

specificity_gbm = cm_gbm[1,1] / (cm_gbm[0,0] + cm_gbm[1,1])
print('Specificity (GBM)' , specificity_gbm)

```
