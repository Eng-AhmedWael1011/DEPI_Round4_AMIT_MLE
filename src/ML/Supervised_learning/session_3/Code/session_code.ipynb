# Lasso basics and how it affects the training data score in linear regression


```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge , Lasso
from sklearn.metrics import mean_squared_error , root_mean_squared_error , r2_score
```

```python
x = np.array([[1],[2],[3],[4]])
y = np.array([2,4,6,8])
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x , y)
y_pred_lin = lin_reg.predict(x)

lin_reg.coef_
lin_reg.intercept_
```

```python
lin_ridge = Ridge(10)
lin_ridge.fit(x , y)
y_pred_lin_Ridge = lin_ridge.predict(x)

print(lin_ridge.coef_)
print(lin_ridge.intercept_)
```

```python
lin_lasso = Lasso(10)
lin_lasso.fit(x , y)
y_pred_lin_lasso = lin_lasso.predict(x)

print(lin_lasso.coef_)
print(lin_lasso.intercept_)
```

```python
print("Linear Regression")
print("MSE" , mean_squared_error(y , y_pred_lin))
print("R2" , r2_score(y , y_pred_lin))
print('_'*20)
print("Lasso Regression")
print("MSE" , mean_squared_error(y , y_pred_lin_lasso))
print("R2" , r2_score(y , y_pred_lin_lasso))
print('_'*20)
print("Ridge Regression")
print("MSE" , mean_squared_error(y , y_pred_lin_Ridge))
print("R2" , r2_score(y , y_pred_lin_Ridge))
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x , y, color='blue' , label = 'Actual Data')
plt.plot(x , y_pred_lin , color='red' , label = 'Linear Regression')
plt.title('LinearRegression')
plt.legend()
plt.show()
# plt.plot(x , y_pred_lin_lasso , color='red' , label = 'Lasso reg')
# plt.plot(x , y_pred_lin_Ridge , color='red' , label = 'ridge reg')
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x , y, color='blue' , label = 'Actual Data')
plt.plot(x , y_pred_lin_lasso , color='red' , label = 'Lasso Regression')
plt.title('Lasso Regression')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.scatter(x , y, color='blue' , label = 'Actual Data')
plt.plot(x , y_pred_lin_Ridge, color='red' , label = 'Ridge Regression')
plt.title('Ridge Regression')
plt.legend()
plt.show()
```

```python
plt.scatter(x , y, color='blue' , label = 'Actual Data')
plt.plot(x , y_pred_lin , color='red' , label = 'Linear Regression')
plt.plot(x , y_pred_lin_lasso ,linestyle="--",  color='cyan' , label = 'Lasso Regression')
plt.plot(x , y_pred_lin_Ridge, color='black' , label = 'Ridge Regression' , linestyle=":")
plt.title('Comparison of linear , Ridge and lasso Regression')
plt.legend()
plt.show()
```
