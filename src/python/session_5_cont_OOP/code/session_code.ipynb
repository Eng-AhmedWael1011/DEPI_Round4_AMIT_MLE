```python
from abc import ABC, abstractmethod


class Animal(ABC):

    @abstractmethod
    def move(self):
        pass


class Bird(Animal):
    def move(self):
        print("bird")

class Cat(Animal):

    def move(self):
        print('cat')

class Dog(Animal):
    pass
```

```python
a = Bird()
```

```python
b = Dog()
```

```python
class A():
    def do_this(self):
        print("I am in A")
class B(A):
    pass
class C():
    def do_this(self):
        print("I am in C")
class D(B,C):
    pass
```

```python
x = D()
x.do_this()
```
