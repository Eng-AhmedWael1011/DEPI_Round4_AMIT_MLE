```python
# Task 1 
# Suppose you have this email address “Amit_ml@gmail.edu”
''' 
- Input Validation: Check if the input string contains exactly one "@" symbol and 
at least one "." after the "@" symbol. If it's not a valid email, return "Invalid 
email". 
 
- Extract Username: Extract and return the part of the email before the "@" 
symbol. 
 
- Extract Domain: Extract and return the domain (the part between "@" and the 
last "."). 
 
- Check for Domain Ending: Check if the email ends with ".com". If it does, 
return "Commercial Domain". If it ends with ".edu", return "Educational Domain". 
Otherwise, return "Other Domain". '''


# Solution 1

# Input 
email = input("Enter Your Email")

# Validation and Output

# check for invalid email
if email.count('@') != 1:
    print("invalid_email")
# if valid -> extract the username
else:
    username = email.split("@")[0]
    domain = email.split("@")[-1]
    print("The username: ",username)
    print("The domain: ",domain)
    # Check if it's a commercial or educational or other
    if domain.__contains__('com'):
        print('Commercial Domain' )
    elif domain.__contains__('edu'):
        print('Educational Domain')
    else:
        print('other domain')
```

```python
# Task-2: 

''' 
Encoded Message: ###!!@mocleW EPGTQ!!!6789 

# Steps to Decode: 

1. Extract the core part of the message: "mocleW EPGTQ". 

2. Reverse the first word: "mocleW" becomes "Welcome". 

3. Replace shifted vowels in the second word: 
o "EPGTQ": No vowels to change. 

4. Final decoded message: "Welcome PGTQ" 

'''
#task 2
# Solution

# Step 1 ->  Extract the sentence
encoded_message =  '###!!@emocleW EPGTQ!!!6789' 
sentence = encoded_message.split('!')[2].replace('@' , '')

# First split to remove the !
# => Output ['###', '', '@emocleW EPGTQ', '', '', '6789']
# [2] to extract the sentence 
# .replace() => to remove the @

# Step 2 -> Split the sentence to reverse the first word the join the again

adjustment = sentence.split(' ')
first_word = adjustment[0]
new_word1 = first_word[::-1]

# step 3 -> remove the Vowels AKA "A,E,O,U,I"
no_vowels = adjustment[1].replace('E' , ' ')

# step 4 -> merge the updated sentence
new_sentence = new_word1 + no_vowels
print(new_sentence)
```

```python
# Task-3

'''

Encoded Message: &&&**$gnirtS PLIO!!@1234 

Steps to Decode: 

1. Extract the core part of the message: "gnirtS PLIO". 

2. Reverse the first word: "gnirtS" becomes "String". 

3. Replace shifted vowels in the second word: 

o "PLIO": Replace I->E and O->U to get "PLEU". 

4. Final decoded message: "String PLEU"


'''

encoded_message =  '&&&**$gnirtS PLIO!!@1234'
sentence = encoded_message.split('!')[0].split("$")[1]
sentence

first_word = sentence.split()[0]
reversed_first_word = first_word[::-1]

second_word = sentence.split()[1]
replaced_word = second_word.replace('I', 'E').replace('O', 'U')

new_sentence = reversed_first_word + " " + replaced_word

print(new_sentence)
```

```python
# Task-4: 

'''

Encoded Message: ##$$$@!yalpstcejorp EPUVT****9887 

Steps to Decode: 

1. Extract the core part of the message: "yalpstcejorp EPUVT". 

2. Reverse the first word: "yalpstcejorp" becomes "projectplay". 

3. Replace shifted vowels in the second word: 

o "EPUVT": Replace E->A, U->O to get "APOVT". 

4. Final decoded message: "projectplay APTOV". 
 



'''

encoded_message =  '##$$$@!yalpstcejorp EPUVT****9887'
sentence = encoded_message.split('!')[1].split("*")[0]
sentence


first_word = sentence.split()[0]
reversed_first_word = first_word[::-1]

second_word = sentence.split()[1]
replaced_word = second_word.replace('E', 'A').replace('U', 'O')

new_sentence = reversed_first_word + " " + replaced_word

print(new_sentence)
```
