```python
def null_handling_strategy(df, strategy="fill_mean"):
    """
    Company Task: Clean a dataset by resolving missing (NaN) values.
    """
    # TODO: Implement using .isnull(), .dropna(), or .fillna() you can used Customer_Age for your test case
    null = df.isnull().sum()
    print(df.head(11))

    print('Before Null handling')

    print(f"the null dataframe {pd.DataFrame({
        'count of null' : null
    }).T}")

    num_cols = df.select_dtypes('number').columns

    if strategy == "drop_rows":
            
            df.dropna(inplace = True)

    elif strategy == "fill_mean": 

        for col in num_cols:

            mean = df[col].mean()
            df[col].fillna(mean , inplace=True)
            
        print("Nulls filled with means")

    elif strategy == "fill_median":   

        for col in num_cols:
            median = df[col].median()
            df[col].fillna(median , inplace=True)

        print("Nulls filled with median")

    else:
        print("no action done as per there is no clear strategy provided")

    print("after Null handling")
    
    print(df.head(11))
    null = df.isnull().sum()
    print(f"the null dataframe {pd.DataFrame({
        'count of null' : null
    }).T}")

    return print(f'Null handled Successfully, strategy used is {strategy}')

    

null_handling_strategy(df=df_test, strategy='fill_median')
```
