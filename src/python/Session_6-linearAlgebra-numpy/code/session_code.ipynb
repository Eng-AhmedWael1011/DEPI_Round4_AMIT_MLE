```python
def array_reshape(arr, rows = -1 , cols = -1):
    try:
        arr = np.array(arr)
        return arr.reshape(rows , cols)
    except (TypeError , ValueError) as e:
        raise ValueError(f"Invalid Reshaped operation, check for the size of the array and try again")
```

```python
array_reshape([1,122,3,43,4,35,25,24,5325,25424] , 2,5)
```
